### STENOSIS WITH BINARY SEGMENTATION MASK IN TEST AND VALIDATION

```python

# Paths for TEST SET
# json_path = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\test\annotations\test.json'
# image_folder = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\test\images'
# path_to_save_folder = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\test\test_binary_segmentation_mask'

# FOR VALIDATION SET
json_path = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\val\annotations\val.json'
image_folder = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\val\images'
path_to_save_folder = r'C:\code_fun\uni ML project\implementation\arcade\arcade\stenosis\val\validation_binary_segmentation_mask'

# Load the JSON file
with open(json_path) as f:
    data = json.load(f)

# Extract image and annotation information
images_info = {img['id']: img for img in data['images']}
annotations = data['annotations']
categories = {cat['id']: cat['name'] for cat in data['categories']}

# Function to create a binary mask from segmentation points
def create_binary_mask(image_shape, segmentation):
    # Create a blank mask (same size as the image but with a single channel)
    mask = np.zeros(image_shape[:2], dtype=np.uint8)

    # Draw the polygon on the mask (fill the region)
    points = np.array(segmentation, dtype=np.int32).reshape((-1, 1, 2))
    cv2.fillPoly(mask, [points], 255)  # Fill with white color (255)

    return mask

# Group annotations by image_id
annotations_by_image = {}
for ann in annotations:
    if ann['image_id'] not in annotations_by_image:
        annotations_by_image[ann['image_id']] = []
    annotations_by_image[ann['image_id']].append(ann)

# Process each image
for image_id, anns in annotations_by_image.items():
    img_info = images_info[image_id]
    img_path = os.path.join(image_folder, img_info['file_name'])
    image = cv2.imread(img_path)

    # Initialize an empty mask for the current image
    mask = np.zeros(image.shape[:2], dtype=np.uint8)

    # Process each annotation for the image and create the binary mask
    for ann in anns:
        for seg in ann['segmentation']:
            # Create a binary mask for each segmentation
            binary_mask = create_binary_mask(image.shape, seg)

            # Combine the binary mask with the existing mask (OR operation)
            mask = cv2.bitwise_or(mask, binary_mask)

    # Save the binary mask in the specified location
    save_path = os.path.join(path_to_save_folder, f"mask_{img_info['file_name'].split('.')[0]}.png")
    cv2.imwrite(save_path, mask)

    print(f"Saved binary mask for image {img_info['file_name']}")
```

# 1- Image prerocessing


```python


# Paths for saving preprocessed data
save_dir = "/content/drive/MyDrive/arcade/stenosis/preprocessed_data"
os.makedirs(save_dir, exist_ok=True)  # Create directory if not exists

# Function to load, preprocess, and save images and masks
def preprocess_and_save(images_path, masks_path, save_name):
    images = []
    masks = []

    image_files = sorted(os.listdir(images_path))
    mask_files = sorted(os.listdir(masks_path))   # agr sorted na kia images and masks ko to training m
                  #  bht issue ay ga, because it is possible that images k 0th index pr 100.png pra ho and masks k
                  # 0th index pr 786.png pra ho

    for img_filename, mask_filename in zip(image_files, mask_files):
        img_path = os.path.join(images_path, img_filename)
        mask_path = os.path.join(masks_path, mask_filename)

        # Read and preprocess the image
        img = cv2.imread(img_path)
        img = cv2.resize(img, (128, 128))
        img = img / 255.0  # scale images also called image normalization, measn scale all values from 0 to 1, we can also do
        # normalization, using normalization function provided by keras.utils as "from tensorflow.keras.utils import normalize
        # normalize(img, axis = 1)"
        images.append(img)

        # Read and preprocess the mask
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        if mask is None:
            print(f"Warning: Mask not found or unreadable for {mask_filename}")
            continue

        mask = cv2.resize(mask, (128, 128))
        mask = np.where(mask > 128, 1, 0)  # Convert to binary
        mask = to_categorical(mask, num_classes=2).squeeze()  #  # I HAVE MADE BINARY SEGMENTATION MASK TO TRAIN THE
                                                #  U-NET MODEL. HERE, to_categorical(mask, num_classes=2) CONVERT
                                                # THE MASK IMAGES (WHICH ARE BINARY) INTO ONE-HOT ENCODING IN
                                                # SUCH A WAY THAT WHERE THE ROI MEANS WHERE THE 1 IS, IT WILL
                                                # CONVERT INTO [0,1] AND THE BACKGROUND REGION MEANS 0 WIL BE
                                                # CONVERTED TO [1,0]. AND I HAVE GIVEN num_classes = 2 BECAUSE
                                                # EACH PIXEL IS DIFFERENTIATED INTO 2 CLASSES EITHER 0 (BG) OR ROI (1).
                                                # converting mask into one-hot encoding is essential because many
                                                # deep learning model don't train properly without this.
                                    # .squeeze() is used further to remove extra dimentions. when images are converted
                                    # into one-hot encoding, to_categorical() add extra dimentions to shape. e.g: if
                                    # before converting to one-hot encoding, the shape is: (HEIGHT, WIDTH, classes),
                                    # after converting to one-hot encdong, the shape becomes (HEIGHT,WIDTH,classes,1)
                                    # so .squeeze() remove this extra dimention 1, .squeeze() removes all
                                    # single dimensions.
        masks.append(mask)

    # Convert lists to NumPy arrays
    images = np.array(images)
    masks = np.array(masks)

    # Save to .npy files
    np.save(os.path.join(save_dir, f"{save_name}_images.npy"), images)
    np.save(os.path.join(save_dir, f"{save_name}_masks.npy"), masks)

    print(f"Saved {len(images)} images and {len(masks)} masks as {save_name}_images.npy and {save_name}_masks.npy")

# Preprocess and save datasets
preprocess_and_save("/content/drive/MyDrive/arcade/stenosis/train/images",
                    "/content/drive/MyDrive/arcade/stenosis/train/binary_segmentation_mask", "train")
preprocess_and_save("/content/drive/MyDrive/arcade/stenosis/val/images",
                    "/content/drive/MyDrive/arcade/stenosis/val/validation_binary_segmentation_mask", "val")
preprocess_and_save("/content/drive/MyDrive/arcade/stenosis/test/images",
                    "/content/drive/MyDrive/arcade/stenosis/test/test_binary_segmentation_mask", "test")
```

```python
import albumentations as A


# Define augmentation pipeline
augment = A.Compose([
    A.HorizontalFlip(p=0.5),  # Flip images and masks horizontally and have 50% chnaces of flipping. yani hm y
    # nhi keh rhy k ap n hr hal m flip krna hi krna h, agr saari hi images flipped hui to model kbhi bhi
    # asal image to dekh hi nhi pay ga, e.g: consider example of car. hm bar bar ulti phulti car hi dekha rhy hn
    # model ko kbhi moqa hi nhi mily ga k vo original seedhi car dekhy, although hm n end m original and augmented
    # images ko join kr dia h, model dekh to le ga k original yani seedhi image esi h but it is good k ap probablity
    # ki value 1.0 yani 100% ki bijay koi diferent value do

    A.VerticalFlip(p=0.5),  # Flip images and masks vertically
    A.RandomRotate90(p=0.5),  # Randomly rotate by 90 degrees
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1, rotate_limit=15, p=0.5), # shift_limit means k max 5%
    # image ko left ya right shift kr do, scale_limit means k max 10 % image ko zoom in ya out kro and rotate_limit
    # here means k 15 degrees image ko rotate kro
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
], additional_targets={'mask': 'mask'})  # Ensure mask gets same transform
```

```python
def augment_data(images, masks, num_augments=9):
    augmented_images = []
    augmented_masks = []

    for i in range(len(images)):  # Loop through each image-mask pair
        for _ in range(num_augments):  # Apply augmentation 'num_augments' times per image
            augmented = augment(image=images[i], mask=masks[i])  # Apply Albumentations
            augmented_images.append(augmented["image"])  # Store augmented image
            augmented_masks.append(augmented["mask"])  # Store augmented mask


    return np.array(augmented_images), np.array(augmented_masks)

# Load original data
train_images = np.load("/content/drive/MyDrive/arcade/stenosis/preprocessed_original_images/train_images.npy")
train_masks = np.load("/content/drive/MyDrive/arcade/stenosis/preprocessed_original_images/train_masks.npy")

# Apply augmentation (each image gets 9 new versions)
aug_train_images, aug_train_masks = augment_data(train_images, train_masks, num_augments=9)

# Combine original and augmented data
train_images_combined = np.concatenate((train_images, aug_train_images))
train_masks_combined = np.concatenate((train_masks, aug_train_masks))

# Save augmented dataset

# os.makedirs("/content/drive/MyDrive/arcade/stenosis/augmented_data", exist_ok=True)

np.save("/content/drive/MyDrive/arcade/stenosis/augmented_data_10K/train_images_10000.npy", train_images_combined)
np.save("/content/drive/MyDrive/arcade/stenosis/augmented_data_10K/train_masks_10000.npy", train_masks_combined)

print(f"Original training images: {train_images.shape}")
print(f"Augmented training images: {aug_train_images.shape}")
print(f"Final training dataset: {train_images_combined.shape}")
```
